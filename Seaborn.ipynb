```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating a sample dataset
np.random.seed(10)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C', 'D'], size=100),
    'Value': np.random.normal(loc=0, scale=1, size=100)
})

# Plotting box plot
sns.boxplot(x='Category', y='Value', data=data)
plt.title('Box Plot')
plt.show()
```

```python
sns.boxplot(x='Value', y='Category', data=data)
plt.title('Box Plot')
plt.show()
```

```python
sns.boxplot(x='Value', y='Category', data=data, width=0.3)
plt.title('Box Plot')
plt.show()
```

```python
# Plotting violin plot
sns.violinplot(x='Category', y='Value', data=data)
plt.title('Violin Plot')
plt.show()
```

```python
# Plotting violin plot
sns.violinplot(x='Value', y='Category', data=data)
plt.title('Violin Plot')
plt.show()
```

```python
sns.swarmplot(x='Category', y='Value', data=data, color='k', size=3)

sns.violinplot(x='Category', y='Value', data=data)
plt.title('Violin Plot')
plt.show()
```

```python
sns.swarmplot(x='Category', y='Value', data=data, color='k', size=3)

sns.lineplot(x='Category', y='Value', data=data)
plt.title('line Plot')
plt.show()
```

```python
sns.swarmplot(x='Category', y='Value', data=data, color='k', size=3)

sns.barplot(x='Category', y='Value', data=data)
plt.title('bar Plot')
plt.show()
```
