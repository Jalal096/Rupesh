```python
import pandas as pd
import numpy as np

# Generate date range
date_range = pd.date_range(start='2020-01-01', end='2022-12-31', freq='M')

# Generate synthetic sales data
sales_data = np.random.randint(10000, 50000, size=len(date_range))

# Create DataFrame
sales_df = pd.DataFrame({'Date': date_range, 'Sales': sales_data})

# Set Date column as index
sales_df.set_index('Date', inplace=True)

print(sales_df.head())
```

```python
import matplotlib.pyplot as plt

# Plot the sales data
plt.figure(figsize=(10, 6))
plt.plot(sales_df.index, sales_df['Sales'], marker='o', linestyle='-')
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
sales_df_log = sales_df.copy() 
```

```python
sales_df_log['Sales'] = np.log(sales_df)
sales_df_log.Sales
```

```python
sales_df_log.plot()
plt.show()
```

```python

# Plot original time series and log-transformed time series
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.title('Original Time Series')
plt.plot(sales_df)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.title('Log Transformed Time Series')
plt.plot(sales_df_log)
plt.xlabel('Date')
plt.ylabel('Log(Sales)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
