```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data3 = {'Year':[2001,2002,2003,2004,2005],
         'Sales':[98,110,170,100,200],
         'City':['Mumbai','Delhi','Kolkata','Ranchi','Patna']}
df=pd.DataFrame(data3)
print(df)
```

```python
plt.scatter(df['Year'],df['Sales'])
plt.title("Scatter Plot")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.plot(df['Year'],df['Sales'],marker='*')
plt.title("Line Plot")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.xticks(df['Year'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.bar(df['City'],df['Sales'])
plt.title("Bar Plot")
plt.xlabel("Cities",fontweight='bold')
plt.ylabel("Sales",fontweight='bold')
plt.grid(True)
plt.show()
```

```python
plt.hist(df['Sales'])
plt.title("Histogram")
plt.show()
```

```python
data3 = {'Year': [2001, 2002, 2003, 2004, 2005],
         'Sales': [98, 110, 170, 100, 200],
         'City': ['Mumbai', 'Delhi', 'Kolkata', 'Ranchi', 'Patna']}
df = pd.DataFrame(data3)
sns.lineplot(x='Year', y='Sales', data=df, label='Sales')
plt.title('Sales Over Year')
plt.xticks(df['Year'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='Year', y='Sales', data=df, label='Sales')
plt.title("Scatter with seaborn")
plt.xticks(df['Year'])
plt.show()
```

```python
sns.barplot(x='City',y='Sales',data=df,label='Sales')
plt.title("Bar with Seaborn")
plt.show()
```

```python
sns.histplot(x='Sales',data=df,label='Sales')
plt.title("Histogram with Seaborn")
plt.show()
```
